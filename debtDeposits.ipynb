```python
import pandas as pd
import json
import numpy as np
```

```python
# from https://www.imf.org/external/datamapper/NFC_LS@GDD/SWE
df = pd.read_csv('raw_data/imf_debt.csv', delimiter='\t', dtype={y: float for y in range(1950,2021)}).replace('no data', np.nan)
display(df)
for col in df.columns[1:]:
    
    df[col] = np.array([float(n) for n in df[col]]) / 100

df.to_csv('debt_ratios.csv')
```

```python
# from https://coronadatasource.org/open-data/geo-data/country-names-codes-all-countries/
country_df = pd.read_csv('raw_data/CountryNamesCodesBasic.csv')

country_map = {}
for i, row in country_df.iterrows():
    country_map[row[0] ]= row[2]

```

```python
country_map["China, People's Republic of"] = country_map["China"]
country_map['Congo, Republic of '] = "DRC"
country_map['Czech Republic'] = country_map['Czechia']
country_map['Hong Kong SAR'] = country_map['Hong Kong']
country_map['Korea, Republic of'] = country_map['South Korea']
country_map['North Macedonia '] = country_map['North Macedonia']
country_map['Russian Federation'] = country_map['Russia']
country_map['Slovak Republic'] = country_map['Slovakia']
country_map['São Tomé and Príncipe'] = country_map['Sao Tome and Principe']
country_map['Türkiye, Republic of'] = country_map['Turkey']
country_map['United States'] = country_map['USA']

df['code'] = [country_map[c] for c in df['Country']]
df = df[['code'] + list(df.columns[:-1])]
df.to_csv('imf_debt_ratios.csv')
df
```
