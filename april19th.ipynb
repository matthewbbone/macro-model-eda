## OECD Investment Matrix

```python
import numpy as np
import pandas as pd
import plotly.express as px
import json
```

$$O_{44x44}=(R_{44x41}V_{41x41}C_{41x44}\quad\emptyset\quad j_{44}R_{44x41}V_{41x41}C_{41x44})$$

Where R sums and splits columns such that rows are split according to the relative weight of total production (e.g. if Agriculture is being split into Farming, which produces 80% and Fishing, which produces 20% of capital). C averages and splits columns such that the averages are weighted according to their total expenditure. We have expenditure weightings of VomLehn-Winberry industries but would need production weightings of OECD industries (some of which we could direcelty map from BEA).

Currently R assumes uniform production when splitting industry rows.
